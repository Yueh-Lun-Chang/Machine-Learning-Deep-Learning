```python
from tensorflow.keras.datasets.fashion_mnist import load_data
((x_train, y_train),(x_test, y_test)) = load_data()
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import random
c = random.randint(0, 59999)
print(y_train[c])
plt.imshow(x_train[c],cmap="gray")
```

```python
import pandas as pd
pd.DataFrame(x_train[c])
```

```python
x_train_norm = x_train.reshape(-1, 784) / 255   
x_test_norm = x_test.reshape(-1, 784) / 255
pd.DataFrame(x_train_norm)
```

```python
from tensorflow.keras.utils import to_categorical
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
print(y_train_cat[0])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

mlp = Sequential()
mlp.add(Dense(128, activation="relu", input_dim=784))
# 128: 可設高一點，留空間給Dropout，但設高也不一定是好事 -> try&error
# 在Dense間加上Dropout
# 做了Dropout訓練結果一定會變好嗎? 不一定，要看訓練資料的特徵特性，但還是要做Dropout以防萬一
# 訓練時只有96個神經元 (128 - 128*0.25)
mlp.add(Dropout(0.25))
mlp.add(Dense(10, activation="softmax"))
mlp.summary()
```

```python
mlp.compile(loss="categorical_crossentropy",
            optimizer="adam",
            metrics=["accuracy"])
```

```python
# callbacks: 每做一次epochs，回調一些功能 
# ex.EarlyStopping:當被監測的數量不再提升，則停止訓練；ModelCheckpoint: 儲存模型
# EarlyStopping(patience=3, restore_best_weights=True)
# patience=0: 連續上升一次就停止

from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
stop_callback = EarlyStopping(patience=3, restore_best_weights=True)
save_callback = ModelCheckpoint("fashion_mnist.h5", save_best_only=True)

mlp.fit(x_train_norm,
        y_train_cat,
        batch_size=200,
        epochs=100,
        validation_split=0.1,
        callbacks=[stop_callback, save_callback],
        verbose=2)
```

```python
pre = mlp.predict_classes(x_test_norm)
pre
```

```python
mlp.evaluate(x_test_norm, y_test_cat)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix

label_list = ["T-shirt/top",
              "Trouser",
              "Pullover",
              "Dress",
              "Coat",
              "Sandal",
              "Shirt",
              "Sneaker",
              "Bag",
              "Ankle boot"]

c = ["{}(預測)".format(i) for i in label_list]
i = ["{}(正確)".format(i) for i in label_list]

mat = confusion_matrix(y_test, pre)
mat = [np.round(row / sum(row), 3) for row in mat]
pd.DataFrame(mat, columns=c, index=i)
```

```python
# [:200]: 太多錯, 我只取前200個
idx = np.nonzero(pre != y_test)[0][:200]

false_pre_img = x_test[idx]
false_pre_label = y_test[idx]
false_pre_predict = pre[idx]

plt.figure(figsize=(14,42))

width = 10
height = int(len(idx) / width) + 1
for (i, img) in enumerate(false_pre_img):
    plt.subplot(height, width, i+1)
    t = "[P]:{}\n[O]:{}".format(label_list[false_pre_predict[i]],
                                label_list[false_pre_label[i]])
    plt.title(t)
    plt.axis("off")
    plt.imshow(img, cmap="gray")
```

```python
from tensorflow.keras.preprocessing.image import load_img
f = input("檔名:")
img_raw = load_img(f, target_size=(28,28)).convert("L")
# .convert("L"): 轉成灰階 ("L"模式為灰階)

# 將圖像攤開並normalize
img = np.array(img_raw).reshape(1, 784) /255
answer = mlp.predict_classes(img)[0]
# 只有一筆資料，因此[0]

print("正確答案應該是:", label_list[answer])
plt.imshow(img_raw, cmap="gray")
```
